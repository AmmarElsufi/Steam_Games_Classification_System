```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv(r"phase_4_data/steam_games_extracted_phase_3.csv")
```

```python
# Log Trasnformation
```

```python
num_data = data.select_dtypes(exclude=["object" , "bool"])

for col in list(num_data.columns) :
    data[col] = np.log(data[col] + 1 ) # adding 1 to avoid log(0)
 
```

```python
# ----------------------------------------------------------------------------------------------------------------------------------
# ----------------------------------------------------------------------------------------------------------------------------------
# ----------------------------------------------------------------------------------------------------------------------------------
```

```python
# Power transformation 
```

```python
from sklearn.preprocessing import PowerTransformer 

pt = PowerTransformer(method="yeo-johnson" , standardize = False)

num_data = data.select_dtypes(exclude=["object" , "bool"])
pt.fit(num_data)
transformed_num_data = pt.transform(num_data)

data[num_data.columns] =  transformed_num_data
```

```python
print(pt.lambdas_)
```

```python
# ----------------------------------------------------------------------------------------------------------------------------------
# ----------------------------------------------------------------------------------------------------------------------------------
# ----------------------------------------------------------------------------------------------------------------------------------
```

```python
# Standardization
```

```python
num_data = data.select_dtypes(exclude=["object" , "bool"])

for col in list(num_data.columns) :
    data[col] = ( data[col] - np.mean(data[col]) ) / np.std(data[col]) 

```

```python
# ----------------------------------------------------------------------------------------------------------------------------------
# ----------------------------------------------------------------------------------------------------------------------------------
# ----------------------------------------------------------------------------------------------------------------------------------
```

```python
# Normalization
```

```python
num_data = data.select_dtypes(exclude=["object" , "bool"])

for col in list(num_data.columns) :
    data[col] = ( data[col] - np.min(data[col]) ) / ( np.max(data[col]) - np.min(data[col]) )

```
