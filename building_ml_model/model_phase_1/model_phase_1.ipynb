```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
######################################################################################################################################
######################################################################################################################################
######################################################################################################################################
```

```python

# Entering and Checking on the data
```

```python
data = pd.read_csv(r"original_data/steam_games_extracted_phase_4.csv")
```

```python
print(data.info() , data.shape  , sep="\n\n\n")
```

```python
data.drop(["date_release"] , axis=1 , inplace=True)
```

```python
data.sample(5)
```

```python
data.dropna(inplace=True)

print(data.isnull().sum() ,end="\n\n\n")
```

```python
######################################################################################################################################
######################################################################################################################################
######################################################################################################################################
```

```python

# Taking a small sample of the data for tests
```

```python
sample_size = 0.1 ## 10%

sampled_data = data.sample(frac=sample_size , random_state=42).reset_index(drop=True)

print(sampled_data.shape , end="\n\n\n")

sampled_data.head(5)
```

```python
######################################################################################################################################
######################################################################################################################################
######################################################################################################################################
```

```python

# Splitting 
```

```python
x_sampled = sampled_data.drop(["steam_ranking"] ,axis=1)
y_sampled = sampled_data["steam_ranking"]
```

```python
from sklearn.model_selection import train_test_split

xs_train , xs_test , ys_train , ys_test = train_test_split(x_sampled , 
                                                           y_sampled , 
                                                           train_size = 0.8 , 
                                                          shuffle = True , 
                                                          random_state = 42 , 
                                                          stratify=y_sampled)
```

```python
######################################################################################################################################
######################################################################################################################################
######################################################################################################################################
```

```python

# Encoding Categorical variables 
```

```python
data.info()
```

```python
data.head(5)
```

```python
columns_mapping = {"Mostly Negative" : 2 , 
                  "Mixed" : 5 , 
                  "Mostly Positive" : 7 , 
                  "Positive" : 8 , 
                  "Very Positive" : 9 , 
                  "Overwhelmingly Positive" : 10 }

ys_train = ys_train.map(columns_mapping)
ys_test = ys_test.map(columns_mapping)
```

```python
col_freq = xs_train["game_title"].value_counts().to_dict()
xs_train["game_title"] = xs_train["game_title"].map(col_freq)
xs_test["game_title"] = xs_test["game_title"].map(col_freq).fillna(0)

# scalling  data
max = np.max(xs_train["game_title"])
min = np.min(xs_train["game_title"])
xs_train["game_title"] = (xs_train["game_title"] - min) / (max - min)
xs_test["game_title"] = (xs_test["game_title"] - min) / (max - min)
```

```python
xs_train["release_date"] = pd.to_datetime(xs_train["release_date"])
reference_date  = xs_train["release_date"].max()
xs_train["release_date_days_since"] = ( reference_date  - xs_train["release_date"] ).dt.days 
xs_train.drop(["release_date"] , axis=1 , inplace=True)

xs_test["release_date"] = pd.to_datetime(xs_test["release_date"])
reference_date  = xs_test["release_date"].max()
xs_test["release_date_days_since"] = ( reference_date  - xs_test["release_date"] ).dt.days 
xs_test.drop(["release_date"] , axis=1 , inplace=True)

# scalling  data
max = np.max(xs_train["release_date_days_since"])
min = np.min(xs_train["release_date_days_since"])
xs_train["release_date_days_since"] = (xs_train["release_date_days_since"] - min) / (max - min)
xs_test["release_date_days_since"] = (xs_test["release_date_days_since"] - min) / (max - min)
```

```python
######################################################################################################################################
######################################################################################################################################
######################################################################################################################################
```

```python

# Choosing the optimal model
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

models = {
    "logistic_regression" : LogisticRegression() , 
    "naive_bayes_gaussian" : GaussianNB() ,
    "naive_bayes_mutinomial" : MultinomialNB() , 
    "knn" : KNeighborsClassifier()  ,
    "decision_trees" : DecisionTreeClassifier() ,
    "svm" : SVC()   }


from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score

for name , model in models.items() :
    skf = StratifiedKFold(n_splits=10 , shuffle=True , random_state=42)
    scores = cross_val_score(model ,xs_train , ys_train ,cv=skf , scoring="f1_macro"  )

    model_scores_mean = np.mean(scores)
    result = f"the model is : {name} ,\nF1 accuracy: {model_scores_mean} \n\n"
    print(result)





```
